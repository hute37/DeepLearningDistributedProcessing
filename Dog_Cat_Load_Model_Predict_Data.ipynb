```python
from keras.models import Sequential
from keras.layers import Flatten, Dense
from keras.optimizers import RMSprop
from keras.applications.vgg16 import VGG16
from keras.models import load_model
# model = load_model('my_dog_cat_model_2.h5')
model = load_model('vgg16_dogcat_transfer.h5')
```

```python
from keras.preprocessing import image
import numpy as np
def read_image(file_path):
    img = image.load_img(file_path, grayscale=False, target_size=(150,150))
    return image.img_to_array(img)/255    
# predict out put
in_data = np.ndarray((1, 150, 150, 3), dtype=np.float32)                
import matplotlib.pyplot as plt
i = np.random.randint(50)
img = read_image('../data/test/'+str(i)+'.jpg')
in_data[0] = img
import matplotlib.pyplot as plt
plt.imshow(img)
plt.show()
predict_label = model.predict(in_data)
print(predict_label)
if predict_label > 0.5:
    print("Dog")
else:
    print("Cat")
```
